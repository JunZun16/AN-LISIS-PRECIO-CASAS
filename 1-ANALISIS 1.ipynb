RELACIONES CATEGORICAS

    CALIDAD DE CASA

        Grafica de caja y bigotes de la calidad de la casa frente al precio de venta

```python
var="OverallQual"
data=pd.concat([datos["SalePrice"],datos[var]], axis=1)
```

```python
f, ax=plt.subplots(figsize=(8,6))
fig=sns.boxplot(x=var,y="SalePrice",data=data)
fig.axis(ymin=0,ymax=800000)
```

    AÑO DE CONSTRUCCION

        Grafica del precio de venta frente al año de construccion

```python
var="YearBuilt"
data=pd.concat([datos["SalePrice"],datos[var]], axis=1)
f, ax=plt.subplots(figsize=(20,8))
fig=sns.boxplot(x=var,y="SalePrice",data=data)
fig.axis(ymin=0,ymax=800000)
plt.xticks(rotation=90)
```

    MATRIZ DE CORRELACION 

    Grafica de correlacion donde se puede visualizar cuales variables influyen mas frente al precio de venta

```python
corrmat = datos.corr(numeric_only = [False])
```

```python
f,ax=plt.subplots(figsize=(12,9))
sns.heatmap(corrmat,vmax=.8,square=True)
```

Observamos que las vairables de Area de vivienda, calidad, area del garage, tamaño del garaje son las variables que mas modifican influyen en el precio de venta.
